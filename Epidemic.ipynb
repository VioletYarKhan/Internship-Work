```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Pop = np.array((1, 0, 0, 0), dtype='float')

# Defining translation matrix
translator = np.array([[0.95, 0.04, 0, 0],
                       [0.05, 0.85, 0, 0],
                       [0.00, 0.10, 1, 0],
                       [0.00, 0.01, 0, 1]], dtype = 'float')

times = int(input('How many cycles do you want to run? '))

AllPops = np.zeros((times+1,4), dtype='float')
AllPops[0] = Pop

for t in range(1, times + 1):
    NewPop = np.dot(translator, AllPops[t-1])   
    AllPops[t] = NewPop  
```

```python
plt.plot(AllPops)
plt.show()
print(AllPops[-1])
```
